üõû "PART 2: Tuple Unpacking + Î¶¨Ïä§Ìä∏ Ï°∞Ìï©"

```python
# Ïó¨Îü¨ ÏÑºÏÑú Îç∞Ïù¥ÌÑ∞Î•º Î¶¨Ïä§Ìä∏Î°ú Í¥ÄÎ¶¨

sensor_readings = [
    (8.5, 45, 'lidar'),      # Í±∞Î¶¨, Í∞ÅÎèÑ, ÏÑºÏÑúÌÉÄÏûÖ
    (12.3, 0, 'camera'),     # Í±∞Î¶¨, Í∞ÅÎèÑ, ÏÑºÏÑúÌÉÄÏûÖ
    (15.7, -30, 'radar'),    # Í±∞Î¶¨, Í∞ÅÎèÑ, ÏÑºÏÑúÌÉÄÏûÖ
    (2.1, 90, 'ultrasonic')  # Í±∞Î¶¨, Í∞ÅÎèÑ, ÏÑºÏÑúÌÉÄÏûÖ
]

print("ÏÑºÏÑúÎ≥Ñ Í∞êÏßÄ Ï†ïÎ≥¥:")
for sensor_data in sensor_readings:
    dist, ang, sensor_type = sensor_data  # tuple unpacking
    print(f"{sensor_type}: {dist}m, {ang}ÎèÑ")


# ÏúÑÌóò ÏÑºÏÑú Ï∞æÍ∏∞ (5m Ïù¥ÎÇ¥)
print("\nÏúÑÌóò Í∞êÏßÄ ÏÑºÏÑú:")
for sensor_data in sensor_readings:
    dist, ang, sensor_type = sensor_data
    if dist < 5.0:
        print(f"‚ö†Ô∏è {sensor_type}: {dist}m (ÏúÑÌóò)")
# Í≤ΩÎ°ú Ìè¨Ïù∏Ìä∏Îì§ÏùÑ Î¶¨Ïä§Ìä∏Î°ú Í¥ÄÎ¶¨
waypoints = [
    (0, 0, 'Ï∂úÎ∞úÏßÄ'),
    (5, 3, 'Í≤ΩÏú†ÏßÄ1'),
    (10, 8, 'Í≤ΩÏú†ÏßÄ2'),
    (15, 12, 'Î™©Ï†ÅÏßÄ')
]

print("\nÍ≤ΩÎ°ú Í≥ÑÌöç:")
for point in waypoints:
    x, y, name = point
    print(f"{name}: ({x}, {y})")
# ÍµêÏ∞®Î°ú Ï†ïÎ≥¥Îì§
intersections = [
    ('ÏÑúÏö∏Ïó≠', 'green', 25),
    ('ÏãúÏ≤≠', 'red', 40),
    ('Ï¢ÖÎ°ú3Í∞Ä', 'yellow', 5),
    ('ÏùÑÏßÄÎ°ú', 'green', 15)
]

print("\nÍµêÏ∞®Î°ú Ïã†Ìò∏Îì± ÌòÑÌô©:")
for intersection in intersections:
    name, signal_color, remaining_time = intersection
    print(f"{name}: {signal_color} ({remaining_time}Ï¥à)")

print("\n" + "="*50)
```

- ÏÖã: processed_intersections - Ï≤òÎ¶¨Îêú ÍµêÏ∞®Î°ú Ï∂îÏ†Å
- forÎ¨∏: ÍµêÏ∞®Î°ú Îç∞Ïù¥ÌÑ∞ ÏàúÌöå

üõû PART 3: Tuple Unpacking + ÎîïÏÖîÎÑàÎ¶¨ Ï°∞Ìï©

```python
print("=== PART 3: Tuple Unpacking + ÎîïÏÖîÎÑàÎ¶¨ Ï°∞Ìï© ===")

# Ï∞®ÎüâÎ≥Ñ ÏúÑÏπò Ï†ïÎ≥¥Î•º ÎîïÏÖîÎÑàÎ¶¨Î°ú Í¥ÄÎ¶¨
vehicle_locations = {
    'car_001': (127.1, 37.5, 60),    # Í≤ΩÎèÑ, ÏúÑÎèÑ, ÏÜçÎèÑ
    'car_002': (127.2, 37.6, 45),
    'car_003': (127.3, 37.4, 80),
    'truck_001': (127.0, 37.3, 70)
}

print("Ï∞®ÎüâÎ≥Ñ ÌòÑÏû¨ ÏúÑÏπò:")
for vehicle_id, location_data in vehicle_locations.items():
    lon, lat, spd = location_data  # tuple unpacking
    print(f"{vehicle_id}: ({lon}, {lat}), {spd}km/h")

# Ï∂©Ï†ÑÏÜå Ï†ïÎ≥¥Î•º ÎîïÏÖîÎÑàÎ¶¨Î°ú Í¥ÄÎ¶¨
charging_stations = {
    'Í∞ïÎÇ®Ï∂©Ï†ÑÏÜå': (127.027, 37.498, 5, 2),    # Í≤ΩÎèÑ, ÏúÑÎèÑ, Ï¥ùÏä¨Î°Ø, ÏÇ¨Ïö©Ï§ëÏä¨Î°Ø
    'ÏÑúÏ¥àÏ∂©Ï†ÑÏÜå': (127.032, 37.483, 8, 6),
    'ÏÜ°ÌååÏ∂©Ï†ÑÏÜå': (127.105, 37.514, 6, 1)
}

print("\nÏ∂©Ï†ÑÏÜå ÌòÑÌô©:")
for station_name, station_info in charging_stations.items():
    lon, lat, total_slots, used_slots = station_info
    available = total_slots - used_slots
    print(f"{station_name}: ÏúÑÏπò({lon}, {lat}), ÏÇ¨Ïö©Í∞ÄÎä• {available}/{total_slots}Ïä¨Î°Ø")

# Ï£ºÏ∞®Ïû• Íµ¨Ïó≠Î≥Ñ Ï†ïÎ≥¥
parking_areas = {
    'AÍµ¨Ïó≠': (50, 45, 'compact'),     # Ï¥ùÏûêÎ¶¨, ÏÇ¨Ïö©Ï§ë, Ï∞®ÎüâÌÅ¨Í∏∞
    'BÍµ¨Ïó≠': (30, 30, 'standard'),
    'CÍµ¨Ïó≠': (20, 15, 'large'),
    'DÍµ¨Ïó≠': (15, 8, 'handicapped')
}

print("\nÏ£ºÏ∞®Ïû• ÌòÑÌô©:")
for area_name, area_info in parking_areas.items():
    total, occupied, car_type = area_info
    available = total - occupied
    utilization = round((occupied / total) * 100, 1)
    print(f"{area_name} ({car_type}): {available}/{total}ÏûêÎ¶¨ ÏÇ¨Ïö©Í∞ÄÎä• (ÏÇ¨Ïö©Î•† {utilization}%)")

print("\n" + "="*50)
```

- ÌäúÌîå: ('EV001', 25, (127.1, 37.5), 'sedan') - Ï∞®Îüâ Ï†ïÎ≥¥
- Î¶¨Ïä§Ìä∏: electric_vehicles - Ï∞®Îüâ Î™©Î°ù
- ÎîïÏÖîÎÑàÎ¶¨: charging_stations - Ï∂©Ï†ÑÏÜå Ï†ïÎ≥¥
- ÏÖã: urgent_vehicles - Í∏¥Í∏â Ï∂©Ï†Ñ Ï∞®Îüâ
- forÎ¨∏: Ï∞®ÎüâÎ≥Ñ, Ï∂©Ï†ÑÏÜåÎ≥Ñ ÏàúÌöå

üõû PART 4: Ïã§Ï†Ñ ÏùëÏö© - Ï¢ÖÌï© ÍµêÌÜµ Í¥ÄÏ†ú ÏãúÏä§ÌÖú

```python
print("=== PART 4: Ïã§Ï†Ñ ÏùëÏö© - Ï¢ÖÌï© ÍµêÌÜµ Í¥ÄÏ†ú ÏãúÏä§ÌÖú ===")

# Ïã§ÏãúÍ∞Ñ ÍµêÌÜµ Îç∞Ïù¥ÌÑ∞ (ÏãúÍ∞Ñ, ÏúÑÏπò(Í≤ΩÎèÑ, ÏúÑÎèÑ), ÏÜçÎèÑ, Ï∞®ÎüâÌÉÄÏûÖ, Ï∞®ÏÑ†)
traffic_data = [
    ('09:15', (127.1, 37.5), 45, 'sedan', 1),   # sedan ‚Üí ÏÑ∏Îã®
    ('09:16', (127.2, 37.6), 60, 'truck', 2),    # truck ‚Üí Ìä∏Îü≠
    ('09:17', (127.3, 37.4), 30, 'bus', 1),      # bus ‚Üí Î≤ÑÏä§
    ('09:18', (127.0, 37.3), 80, 'sedan', 3)     # sedan ‚Üí ÏÑ∏Îã®
]

print("Ïã§ÏãúÍ∞Ñ ÍµêÌÜµ Î™®ÎãàÌÑ∞ÎßÅ:")  # Ïã§ÏãúÍ∞Ñ ÍµêÌÜµ ÏÉÅÌô© Ï∂úÎ†•
for data in traffic_data:
    time, position, speed, vehicle_type, lane = data  # ÌäúÌîå Í∞í Ïñ∏Ìå®ÌÇπ (ÏãúÍ∞Ñ, ÏúÑÏπò, ÏÜçÎèÑ, Ï∞®ÎüâÏú†Ìòï, Ï∞®ÏÑ†)
    lon, lat = position  # ÏúÑÏπò Ï†ïÎ≥¥ÎèÑ Îî∞Î°ú ÎÇòÎàî (Í≤ΩÎèÑ, ÏúÑÎèÑ)
    print(f"{time} | {vehicle_type} | ({lon}, {lat}) | {speed}km/h | {lane}Ï∞®ÏÑ†")
    # Ïòà: 09:15 | sedan | (127.1, 37.5) | 45km/h | 1Ï∞®ÏÑ†

# ÏÇ¨Í≥† Ïã†Í≥† Îç∞Ïù¥ÌÑ∞ (ÏãúÍ∞Ñ, ÏúÑÏπò, Ïã¨Í∞ÅÎèÑ, Ï≤òÎ¶¨ Ïó¨Î∂Ä)
accident_reports = [
    ('10:30', (127.15, 37.55), 'minor', False),  # minor ‚Üí Í≤ΩÎØ∏Ìïú ÏÇ¨Í≥†, False ‚Üí ÏïÑÏßÅ Ï≤òÎ¶¨ Ïïà Îê®
    ('10:45', (127.25, 37.65), 'major', True),   # major ‚Üí ÌÅ∞ ÏÇ¨Í≥†, True ‚Üí Ï≤òÎ¶¨ ÏôÑÎ£å
    ('11:00', (127.35, 37.45), 'minor', False)   # Ï≤òÎ¶¨ Ïïà Îêú Í≤ΩÎØ∏Ìïú ÏÇ¨Í≥†
]

print("\nÏÇ¨Í≥† Ïã†Í≥† ÌòÑÌô©:")  # Ï§ÑÎ∞îÍøà ÌõÑ ÏÇ¨Í≥† Ï†ïÎ≥¥ Ï∂úÎ†•
for report in accident_reports:
    time, location, severity, handled = report  # Îç∞Ïù¥ÌÑ∞ Ïñ∏Ìå®ÌÇπ
    lon, lat = location
    status = "Ï≤òÎ¶¨ÏôÑÎ£å" if handled else "Ï≤òÎ¶¨Ï§ë"  # handled: True ‚Üí ÏôÑÎ£å, False ‚Üí ÏßÑÌñâ Ï§ë
    print(f"{time} | ({lon}, {lat}) | {severity} | {status}")
    # Ïòà: 10:45 | (127.25, 37.65) | major | Ï≤òÎ¶¨ÏôÑÎ£å

# Ï≤òÎ¶¨ÎêòÏßÄ ÏïäÏùÄ ÏÇ¨Í≥†Îßå Ï∂úÎ†•
print("\nÎØ∏Ï≤òÎ¶¨ ÏÇ¨Í≥†:")
for report in accident_reports:
    time, location, severity, handled = report
    if not handled:  # Ï≤òÎ¶¨ Ïïà Îêú Í≤ΩÏö∞Îßå
        lon, lat = location
        print(f"‚ö†Ô∏è {time} | ({lon}, {lat}) | {severity} - Ï¶âÏãú Ï≤òÎ¶¨ ÌïÑÏöî")
        # ‚ö†Ô∏è Í≤ΩÍ≥† ÌëúÏãú Ìè¨Ìï®

# ÎÇ†Ïî®Î≥Ñ Í∂åÏû• Ï£ºÌñâ Ï°∞Í±¥ (ÎÇ†Ïî®, ÏµúÎåÄÏÜçÎèÑ, ÏïàÏ†ÑÍ±∞Î¶¨ Î∞∞Ïàò, ÎùºÏù¥Ìä∏ ÏÇ¨Ïö© Ïó¨Î∂Ä)
weather_conditions = [
    ('ÎßëÏùå', 100, 1.0, False),  # False ‚Üí ÎùºÏù¥Ìä∏ ÏÇ¨Ïö© Ïïà Ìï¥ÎèÑ Îê®
    ('ÎπÑ', 80, 1.5, True),     # True ‚Üí ÎùºÏù¥Ìä∏ ÏºúÏïº Ìï®
    ('Îàà', 60, 2.0, True),
    ('ÏïàÍ∞ú', 50, 2.5, True)
]

current_weather = 'ÎπÑ'  # ÌòÑÏû¨ ÎÇ†Ïî® ÏÑ§Ï†ï ('ÎßëÏùå', 'ÎπÑ', 'Îàà', 'ÏïàÍ∞ú' Ï§ë ÌïòÎÇò)

print(f"\nÌòÑÏû¨ ÎÇ†Ïî®: {current_weather}")
print("Í∂åÏû• Ï£ºÌñâ Ï°∞Í±¥:")
for condition in weather_conditions:
    weather, max_speed, distance_factor, lights_on = condition
    if weather == current_weather:
        print(f"ÏµúÎåÄÏÜçÎèÑ: {max_speed}km/h")  # max_speed ‚Üí speed limit in km/h
        print(f"ÏïàÏ†ÑÍ±∞Î¶¨: ÌèâÏÜåÏùò {distance_factor}Î∞∞")  # distance_factor ‚Üí following distance multiplier
        print(f"ÎùºÏù¥Ìä∏ ÏÇ¨Ïö©: {'ÌïÑÏàò' if lights_on else 'ÏÑ†ÌÉù'}")  # lights_on ‚Üí whether headlights are required
        break  # ÌòÑÏû¨ ÎÇ†Ïî® Ï°∞Í±¥ Ï∞æÏïòÏúºÎ©¥ Ï¢ÖÎ£å

# Ï∞®Îüâ ÏÉÅÌÉú ÏßÑÎã® (Ï∞®ÎüâID, ÏóîÏßÑ ÏÉÅÌÉú Ï†êÏàò, Î∏åÎ†àÏù¥ÌÅ¨ ÏÉÅÌÉú Ï†êÏàò, Ï¢ÖÌï© ÌåêÏ†ï)
vehicle_diagnostics = {
    'EV001': (95, 88, 'excellent'),    # excellent ‚Üí Îß§Ïö∞ ÏñëÌò∏
    'EV002': (82, 75, 'good'),         # good ‚Üí ÏñëÌò∏
    'EV003': (65, 45, 'caution'),      # caution ‚Üí Ï£ºÏùò ÌïÑÏöî
    'EV004': (45, 30, 'maintenance')   # maintenance ‚Üí Ï†ïÎπÑ ÌïÑÏöî
}

print("\nÏ∞®Îüâ ÏÉÅÌÉú ÏßÑÎã®:")
for vehicle_id, diagnosis in vehicle_diagnostics.items():
    engine_score, brake_score, overall_status = diagnosis
    avg_score = (engine_score + brake_score) / 2  # ÌèâÍ∑† Ï†êÏàò Í≥ÑÏÇ∞
    print(f"{vehicle_id}: ÏóîÏßÑ {engine_score}Ï†ê, Î∏åÎ†àÏù¥ÌÅ¨ {brake_score}Ï†ê, ÌèâÍ∑† {avg_score:.1f}Ï†ê ({overall_status})")
    # overall_statusÎäî 'excellent', 'good', 'caution', 'maintenance'

# Ï†ïÎπÑÍ∞Ä ÌïÑÏöîÌïú Ï∞®ÎüâÎßå Îî∞Î°ú Ï∂úÎ†•
print("\nÏ†ïÎπÑ ÌïÑÏöî Ï∞®Îüâ:")
for vehicle_id, diagnosis in vehicle_diagnostics.items():
    engine_score, brake_score, overall_status = diagnosis
    if overall_status in ['caution', 'maintenance']:  # Ï£ºÏùò ÎòêÎäî Ï†ïÎπÑ ÌïÑÏöî ÏÉÅÌÉúÎßå Í±∏Îü¨ÎÉÑ
        print(f"üîß {vehicle_id}: {overall_status} - Ï†êÍ≤Ä ÌïÑÏöî")
```
