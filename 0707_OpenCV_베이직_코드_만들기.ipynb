```python
# OpenCV Ï¥àÎ≥¥Ïûê ÌÖåÏä§Ìä∏ ÏΩîÎìú - ColabÏóêÏÑú Ïã§ÌñâÌïòÏÑ∏Ïöî!

import cv2  #OpenCV ÎùºÏù¥Î∏åÎü¨Î¶¨Î•º Î∂àÎü¨ÏòµÎãàÎã§. Ïù¥ÎØ∏ÏßÄ, ÎπÑÎîîÏò§ Ï≤òÎ¶¨ Îì±Ïóê ÎÑêÎ¶¨ ÏÇ¨Ïö©.
import numpy as np  #NumPy ÎùºÏù¥Î∏åÎü¨Î¶¨Î•º Î∂àÎü¨Ïò§Í≥†, Î≥ÑÏπ≠ÏúºÎ°ú npÎ•º ÏÇ¨Ïö©Ìï©ÎãàÎã§. ÏàòÏπòÏó∞ÏÇ∞, ÌñâÎ†¨ Î∞è Î∞∞Ïó¥ Ïó∞ÏÇ∞, Ïù¥ÎØ∏ÏßÄ Îç∞Ïù¥ÌÑ∞Î•º Î∞∞Ïó¥Î°ú Îã§Î£®ÎäîÎç∞ ÌïÑÏàòÏ†ÅÏù¥Îã§.
import matplotlib.pyplot as plt #MatplotlibÏùò pyplot Î™®ÎìàÏùÑ Î∂àÎü¨Ïò§Í≥†, Î≥ÑÏπ≠ÏúºÎ°ú pltÎ•º ÏÇ¨Ïö©Ìï©ÎãàÎã§. Îç∞Ïù¥ÌÑ∞ÏãúÍ∞ÅÌôî(Í∑∏ÎûòÌîÑ, Ïù¥ÎØ∏ÏßÄ)Îì±Ïóê ÏÇ¨Ïö©
from google.colab import files #Google Colab ÌôòÍ≤ΩÏóêÏÑú ÌååÏùºÏùÑ ÏóÖÎ°úÎìúÌïòÍ±∞ÎÇò Îã§Ïö¥Î°úÎìúÌï† Îïå ÏÇ¨Ïö©ÌïòÎäî Î™®ÎìàÏûÖÎãàÎã§.
                               #ÏÇ¨Ïö©ÏûêÍ∞Ä ÏûêÏã†Ïùò Ïª¥Ìì®ÌÑ∞ÏóêÏÑú ColabÏúºÎ°ú ÌååÏùºÏùÑ Ïò¨Î¶¨Í±∞ÎÇò, ColabÏóêÏÑú ÌååÏùºÏùÑ ÎÇ¥Î†§Î∞õÏùÑ Îïå ÌôúÏö©
import io  #ÌååÏù¥Ïç¨ ÎÇ¥Ïû• Î™®ÎìàÏù∏ ioÎäî Î©îÎ™®Î¶¨ ÎÇ¥ÏóêÏÑú ÌååÏùº Í∞ùÏ≤¥(Î∞îÏù¥Ìä∏ Ïä§Ìä∏Î¶º Îì±)Î•º Îã§Î£∞ Îïå ÏÇ¨Ïö©.
           #ÏòàÎ•º Îì§Ïñ¥, ÏóÖÎ°úÎìúÎêú ÌååÏùºÏùò Îç∞Ïù¥ÌÑ∞Î•º Î©îÎ™®Î¶¨ÏóêÏÑú Î∞îÎ°ú ÏùΩÏñ¥Ïò¨ Îïå ÌôúÏö©Îê©ÎãàÎã§
from PIL import Image   #Pillow ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò Image ÌÅ¥ÎûòÏä§Î•º Î∂àÎü¨ÏòµÎãàÎã§.
                        #Ïù¥ÎØ∏ÏßÄÎ•º Ïó¥Í≥†, Ï†ÄÏû•ÌïòÍ≥†, Î≥ÄÌôòÌïòÎäî Îì± Îã§ÏñëÌïú Ïù¥ÎØ∏ÏßÄ ÏûëÏóÖÏùÑ Ìï† Ïàò ÏûàÏäµÎãàÎã§.

print(f"üéâ OpenCV Î≤ÑÏ†Ñ: {cv2.__version__}")

# =============================================================================
# üìù ÌÖåÏä§Ìä∏ 1: Í∞ÑÎã®Ìïú Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ±ÌïòÍ∏∞
# =============================================================================

def test1_create_image():
    """Í∞ÑÎã®Ìïú Ïù¥ÎØ∏ÏßÄ ÎßåÎì§Ïñ¥Î≥¥Í∏∞"""
    print("\nüé® ÌÖåÏä§Ìä∏ 1: Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ±ÌïòÍ∏∞")

    # 1. Îπà Ï∫îÎ≤ÑÏä§ ÎßåÎì§Í∏∞ (Í≤ÄÏùÄÏÉâ)
    black_img = np.zeros((200, 300, 3), dtype=np.uint8)
    #np.zerosÎäî Î™®Îì† Í∞íÏù¥ 0Ïù∏ Î∞∞Ïó¥(Ïù¥ÎØ∏ÏßÄ)ÏùÑ ÎßåÎì≠ÎãàÎã§. -> Î™®Îì† ÌîΩÏÖÄÏù¥ 0Ïù¥ÎØÄÎ°ú ÏôÑÏ†ÑÌûà Í≤ÄÏùÄÏÉâ Ïù¥ÎØ∏ÏßÄÍ∞Ä Îê®.
    #200(ÎÜíÏù¥) x 300(ÎÑàÎπÑ) ÌÅ¨Í∏∞Ïùò 3Ï±ÑÎÑê(Ïª¨Îü¨, BGR) Ïù¥ÎØ∏ÏßÄÎ•º ÏÉùÏÑ±

    # 2. Ìù∞ÏÉâ Ï∫îÎ≤ÑÏä§ ÎßåÎì§Í∏∞
    white_img = np.ones((200, 300, 3), dtype=np.uint8) * 255
    #np.onesÎ°ú Î™®Îì† Í∞íÏù¥ 1Ïù∏ Î∞∞Ïó¥ÏùÑ ÎßåÎì§Í≥†, 255Î•º Í≥±Ìï¥ Î™®Îì† ÌîΩÏÖÄ Í∞íÏùÑ 255Î°ú ÎßåÎì≠ÎãàÎã§.(255Îäî Ìù∞ÏÉâ. Í∞ÄÏû• Î∞ùÏùÄÍ∞í)

    # 3. Îπ®Í∞ÑÏÉâ Ï∫îÎ≤ÑÏä§ ÎßåÎì§Í∏∞
    red_img = np.zeros((200, 300, 3), dtype=np.uint8)
    red_img[:, :, 2] = 255  # BGRÏóêÏÑú RÏ±ÑÎÑê
    #Î®ºÏ†Ä Î™®Îì† Í∞íÏù¥ 0Ïù∏ Í≤ÄÏùÄÏÉâ Ïù¥ÎØ∏ÏßÄÎ•º ÎßåÎì§Í≥†, ÎßàÏßÄÎßâ Ï±ÑÎÑê(Îπ®Í∞ÑÏÉâ, R)ÏóêÎßå 255Î•º Ìï†ÎãπÌï©ÎãàÎã§.
    #OpenCVÏóêÏÑúÎäî ÏÉâÏÉÅ ÏàúÏÑúÍ∞Ä BGRÏù¥ÎØÄÎ°ú, [:,:,2]Í∞Ä Îπ®Í∞ÑÏÉâ Ï±ÑÎÑê

    # Í≤∞Í≥º Ï∂úÎ†•
    plt.figure(figsize=(12, 4)) #Ï∂úÎ†•Ìï† Í∑∏Î¶ºÏùò ÌÅ¨Í∏∞Î•º ÏÑ§Ï†ï


    #Í∞Å Ïù¥ÎØ∏ÏßÄÎ•º subplotÏúºÎ°ú ÎÇòÎàÑÏñ¥ Ìïú ÌôîÎ©¥Ïóê Î≥¥Ïó¨Ï§çÎãàÎã§
    plt.subplot(1, 3, 1) #Í≤ÄÏùÄÏÉâ Ïù¥ÎØ∏ÏßÄ ÌëúÏãú
    plt.imshow(black_img)
    plt.title('Black Image')
    plt.axis('off')

    plt.subplot(1, 3, 2) #Ìù∞ÏÉâ Ïù¥ÎØ∏ÏßÄ ÌëúÏãú
    plt.imshow(white_img)
    plt.title('White Image')
    plt.axis('off')

    plt.subplot(1, 3, 3) #Îπ®Í∞ÑÏÉâ Ïù¥ÎØ∏ÏßÄÎäî OpenCVÏùò BGRÏùÑ MatplotlibÏùò RGBÎ°ú Î≥ÄÌôò ÌõÑ ÌëúÏãú
    plt.imshow(cv2.cvtColor(red_img, cv2.COLOR_BGR2RGB))
    plt.title('Red Image')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
    # plt.tight_layout()ÏúºÎ°ú Í∞ÑÍ≤©ÏùÑ Ï°∞Ï†ïÌïòÍ≥†, plt.show()Î°ú Ïù¥ÎØ∏ÏßÄÎ•º ÌôîÎ©¥Ïóê Ï∂úÎ†•

# Ìï®Ïàò Ïã§Ìñâ ÏΩîÎìú Ï∂îÍ∞Ä
if __name__ == "__main__":
    test1_create_image()
    #(ÌòÑÏû¨ ÌååÏùºÏùÑ ÏßÅÏ†ë Ïã§ÌñâÌï† ÎïåÎßå test1_create_image() Ìï®ÏàòÎ•º Ïã§ÌñâÌï©ÎãàÎã§. ÎßåÏïΩ Ïù¥ ÌååÏùºÏùÑ Îã§Î•∏ ÌååÏù¥Ïç¨ ÌååÏùºÏóêÏÑú importÌïòÎ©¥, test1_create_image()Îäî ÏûêÎèôÏúºÎ°ú Ïã§ÌñâÎêòÏßÄ ÏïäÏäµÎãàÎã§.)
```

```python
# =============================================================================
# üìù ÌÖåÏä§Ìä∏ 2: ÎèÑÌòï Í∑∏Î¶¨Í∏∞
# =============================================================================

def test2_draw_shapes(): #Ìï®Ïàò Ï†ïÏùò
    """ÎèÑÌòï Í∑∏Î¶¨Í∏∞ Ïó∞Ïäµ"""
    print("\nüî∫ ÌÖåÏä§Ìä∏ 2: ÎèÑÌòï Í∑∏Î¶¨Í∏∞")

    # Ï∫îÎ≤ÑÏä§ (Ìù∞Î∞îÌÉï) ÏÉùÏÑ±
    canvas = np.ones((400, 600, 3), dtype=np.uint8) * 255 #(255Îäî Ìù∞ÏÉâÏùÑ ÏùòÎØ∏)

    # 1. ÏÇ¨Í∞ÅÌòï Í∑∏Î¶¨Í∏∞
    cv2.rectangle(canvas, (50, 50), (200, 150), (255, 0, 0), 3)
    #(50, 50)ÏóêÏÑú (200, 150)ÍπåÏßÄÏùò ÏòÅÏó≠Ïóê ÌååÎûÄÏÉâ(Blue, BGR: (255, 0, 0)) ÌÖåÎëêÎ¶¨ ÎëêÍªò 3Ïùò ÏÇ¨Í∞ÅÌòïÏùÑ Í∑∏Î¶ΩÎãàÎã§.
    cv2.rectangle(canvas, (250, 50), (400, 150), (0, 255, 0), -1)
    #(250, 50)ÏóêÏÑú (400, 150)ÍπåÏßÄÏùò ÏòÅÏó≠Ïóê Ï¥àÎ°ùÏÉâ(0, 255, 0)ÏúºÎ°ú ÎÇ¥Î∂ÄÍ∞Ä Ï±ÑÏõåÏßÑ ÏÇ¨Í∞ÅÌòïÏùÑ Í∑∏Î¶ΩÎãàÎã§(-1ÏùÄ ÎÇ¥Î∂Ä Ï±ÑÏõÄ).


    # 2. Ïõê Í∑∏Î¶¨Í∏∞
    cv2.circle(canvas, (125, 250), 50, (0, 0, 255), 3)
    #Ï§ëÏã¨Ïù¥ (125, 250), Î∞òÏßÄÎ¶ÑÏù¥ 50Ïù∏ ÏõêÏùÑ Îπ®Í∞ÑÏÉâ(0, 0, 255) ÌÖåÎëêÎ¶¨ ÎëêÍªò 3ÏúºÎ°ú Í∑∏Î¶º
    cv2.circle(canvas, (325, 250), 50, (255, 255, 0), -1)
    #Ï§ëÏã¨Ïù¥ (325, 250), Î∞òÏßÄÎ¶ÑÏù¥ 50Ïù∏ ÏõêÏùÑ ÎÖ∏ÎûÄÏÉâ(255, 255, 0)ÏúºÎ°ú ÎÇ¥Î∂ÄÎ•º Ï±ÑÏõå Í∑∏Î¶º


    # 3. ÏßÅÏÑ† Í∑∏Î¶¨Í∏∞
    cv2.line(canvas, (450, 50), (550, 150), (255, 0, 255), 5)
    #(450, 50)ÏóêÏÑú (550, 150)ÍπåÏßÄ Î≥¥ÎùºÏÉâ(255, 0, 255) ÎëêÍªò 5Ïùò ÏßÅÏÑ†ÏùÑ Í∑∏Î¶º


    # 4. ÌÖçÏä§Ìä∏ Ïì∞Í∏∞
    cv2.putText(canvas, 'Hello OpenCV!', (200, 350),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2)
    #(200, 350) ÏúÑÏπòÏóê 'Hello OpenCV!'ÎùºÎäî Í∏ÄÏûêÎ•º ÏîÅÎãàÎã§.
    #Ìè∞Ìä∏Îäî FONT_HERSHEY_SIMPLEX, ÌÅ¨Í∏∞Îäî 1, ÏÉâÏÉÅÏùÄ Í≤ÄÏ†ï(0, 0, 0), ÎëêÍªòÎäî 2ÏûÖÎãàÎã§.


    # Í≤∞Í≥º Ï∂úÎ†•
    plt.figure(figsize=(10, 6)) #Ï∂úÎ†• Ï∞Ω ÌÅ¨Í∏∞ ÏÑ§Ï†ï
    plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))  #OpenCVÏùò BGR Ïù¥ÎØ∏ÏßÄÎ•º MatplotlibÏóêÏÑú Ïò¨Î∞îÎ•¥Í≤å Î≥¥Ïù¥ÎèÑÎ°ù RGBÎ°ú Î≥ÄÌôòÌï©ÎãàÎã§.
    plt.title('Shape Drawing Test')
    plt.axis('off')
    plt.show()
    #Ï†úÎ™©ÏùÑ Î∂ôÏù¥Í≥†, Ï∂ïÏùÑ Ïà®Í∏¥ Îí§ Ïù¥ÎØ∏ÏßÄÎ•º ÌôîÎ©¥Ïóê ÌëúÏãúÌï©ÎãàÎã§.

# Ìï®Ïàò Ïã§Ìñâ ÏΩîÎìú Ï∂îÍ∞Ä
if __name__ == "__main__":
    test2_draw_shapes()
    #Ïù¥ ÌååÏùºÏùÑ ÏßÅÏ†ë Ïã§ÌñâÌï† ÎïåÎßå test2_draw_shapes() Ìï®ÏàòÍ∞Ä Ïã§Ìñâ
```

```python
# =============================================================================
# üìù ÌÖåÏä§Ìä∏ 3: ÏÉâÏÉÅ Î≥ÄÌôò
# =============================================================================

def test3_color_conversion():
    """ÏÉâÏÉÅ Î≥ÄÌôò ÌÖåÏä§Ìä∏"""
    print("\nüåà ÌÖåÏä§Ìä∏ 3: ÏÉâÏÉÅ Î≥ÄÌôò")

    # Ïª¨Îü¨ Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ± (Î¨¥ÏßÄÍ∞ú Ìå®ÌÑ¥)
    img = np.zeros((200, 600, 3), dtype=np.uint8)
    #200(ÎÜíÏù¥) x 600(ÎÑàÎπÑ) ÌîΩÏÖÄÏùò 3Ï±ÑÎÑê(BGR) Ïª¨Îü¨ Ïù¥ÎØ∏ÏßÄÎ•º ÎßåÎìúÎäîÎç∞, Î™®Îì† Í∞íÏù¥ 0Ïù¥ÎØÄÎ°ú Í∏∞Î≥∏ÏùÄ Í≤ÄÏùÄÏÉâ Î∞îÌÉïÏûÖÎãàÎã§.


    # Î¨¥ÏßÄÍ∞ú ÏÉâÏÉÅ ÎßåÎì§Í∏∞
    colors = [
        (255, 0, 0),    # Îπ®Í∞ï
        (255, 127, 0),  # Ï£ºÌô©
        (255, 255, 0),  # ÎÖ∏Îûë
        (0, 255, 0),    # Ï¥àÎ°ù
        (0, 0, 255),    # ÌååÎûë
        (75, 0, 130),   # ÎÇ®ÏÉâ
        (148, 0, 211)   # Î≥¥Îùº
    ] #Î¨¥ÏßÄÍ∞úÎ•º Íµ¨ÏÑ±ÌïòÎäî 7Í∞ÄÏßÄ ÏÉâÏÉÅÏùò BGRÌäúÌîåÏùÑ Î¶¨Ïä§Ìä∏Î°ú Ï†ÄÏû•

    # Í∞Å ÏÉâÏÉÅÏùÑ ÏÑ∏Î°ú Îù†Î°ú ÎßåÎì§Í∏∞
    strip_width = 600 // len(colors) #Ïù¥ÎØ∏ÏßÄÏùò Ï†ÑÏ≤¥ ÎÑàÎπÑ(600ÌîΩÏÖÄ)Î•º ÏÉâÏÉÅ Í∞úÏàò(len(colors), Ïó¨Í∏∞ÏÑúÎäî 7)Î°ú ÎÇòÎàïÎãàÎã§
    for i, color in enumerate(colors):  #ÏÉâÏÉÅ Î¶¨Ïä§Ìä∏Î•º Ïù∏Îç±Ïä§ÏôÄ Ìï®Íªò Î∞òÎ≥µ
        x_start = i * strip_width       #Í∞Å Îù†Ïùò ÏãúÏûë xÏ¢åÌëúÎ•º Í≥ÑÏÇ∞
        x_end = (i + 1) * strip_width   #Í∞Å Îù†Ïùò ÎÅù xÏ¢åÌëúÎ•º Í≥ÑÏÇ∞
        img[:, x_start:x_end] = color   #Ïù¥ÎØ∏ÏßÄÏùò Ìï¥Îãπ ÏÑ∏Î°ú ÏòÅÏó≠ Ï†ÑÏ≤¥Ïóê ÌòÑÏû¨ ÏÉâÏÉÅÏùÑ Ï±ÑÏö¥Îã§

    # Îã§ÏñëÌïú ÏÉâÍ≥µÍ∞ÑÏúºÎ°ú Î≥ÄÌôò
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #ÌöåÏÉâÏ°∞
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)  #HSV(ÏÉâÏÉÅ/Ï±ÑÎèÑ/Î™ÖÎèÑ)
    lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)  #LAB(Î∞ùÍ∏∞/Ï†ÅÎ°ù/Ï≤≠Ìô©)

    # Í≤∞Í≥º Ï∂úÎ†•
    plt.figure(figsize=(15, 10)) #Ï†ÑÏ≤¥ Ï∂úÎ†• Í∑∏Î¶ºÏùò ÌÅ¨Í∏∞ ÏÑ§Ï†ï

    plt.subplot(2, 2, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Original (RGB)')
    plt.axis('off')
    #ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÎ•º BGRÏóêÏÑú RGBÎ°ú Î≥ÄÌôòÌï¥ Ï∂úÎ†•

    plt.subplot(2, 2, 2)
    plt.imshow(gray, cmap='gray')
    plt.title('Grayscale')
    plt.axis('off')
    #ÌöåÏÉâÏ°∞ Ïù¥ÎØ∏ÏßÄÎ•º Ï∂úÎ†•

    plt.subplot(2, 2, 3)
    plt.imshow(hsv)
    plt.title('HSV')
    plt.axis('off')
    #HSV ÏÉâÍ≥µÍ∞Ñ Ïù¥ÎØ∏ÏßÄÎ•º Ï∂úÎ†•

    plt.subplot(2, 2, 4)
    plt.imshow(lab)
    plt.title('LAB')
    plt.axis('off')
    #LAB ÏÉâÍ≥µÍ∞Ñ Ïù¥ÎØ∏ÏßÄ Ï∂úÎ†•

    plt.tight_layout()
    plt.show()
    #ÏÑúÎ∏åÌîåÎ°ù Í∞ÑÍ≤©ÏùÑ Ï°∞Ï†ïÌïòÍ≥† Ïù¥ÎØ∏ÏßÄÎ•º ÌôîÎ©¥Ïóê ÌëúÏãú

# Ìï®Ïàò Ïã§Ìñâ ÏΩîÎìú Ï∂îÍ∞Ä
if __name__ == "__main__":
    test3_color_conversion()
```

```python
# =============================================================================
# üìù ÌÖåÏä§Ìä∏ 4: Ïù¥ÎØ∏ÏßÄ ÌïÑÌÑ∞ÎßÅ
# =============================================================================

def test4_image_filtering(): #Ìï®Ïàò Ïù¥Î¶Ñ Ï†ïÏùò
    """Ïù¥ÎØ∏ÏßÄ ÌïÑÌÑ∞ÎßÅ ÌÖåÏä§Ìä∏"""
    print("\nüîç ÌÖåÏä§Ìä∏ 4: Ïù¥ÎØ∏ÏßÄ ÌïÑÌÑ∞ÎßÅ")

    # ÎÖ∏Ïù¥Ï¶àÍ∞Ä ÏûàÎäî Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ±
    img = np.ones((300, 300, 3), dtype=np.uint8) * 128  # ÌöåÏÉâ Î∞∞Í≤Ω
    #300x300 ÌÅ¨Í∏∞Ïùò 3Ï±ÑÎÑê(BGR) Ïª¨Îü¨ Ïù¥ÎØ∏ÏßÄÎ•º ÎßåÎì†Îã§. (128: ÌöåÏÉâ)

    # Ïõê Í∑∏Î¶¨Í∏∞
    cv2.circle(img, (150, 150), 80, (255, 255, 255), -1)
    #Ï§ëÏã¨Ïù¥ (150, 150), Î∞òÏßÄÎ¶ÑÏù¥ 80Ïù∏ Ìù∞ÏÉâ(255, 255, 255) ÏõêÏùÑ Ïù¥ÎØ∏ÏßÄÏóê Í∑∏Î¶º. -1ÏùÄ Ïõê ÎÇ¥Î∂ÄÎ•º Ï±ÑÏö¥Îã§Îäî Îúª

    # ÎÖ∏Ïù¥Ï¶à Ï∂îÍ∞Ä
    noise = np.random.randint(0, 50, img.shape, dtype=np.uint8)
    noisy_img = cv2.add(img, noise)
    #0~49 ÏÇ¨Ïù¥Ïùò Î¨¥ÏûëÏúÑ Í∞íÏúºÎ°ú Ïù¥ÎØ∏ÏßÄÏôÄ Í∞ôÏùÄ ÌÅ¨Í∏∞Ïùò ÎÖ∏Ïù¥Ï¶à Î∞∞Ïó¥ÏùÑ ÎßåÎì†Îã§. ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÏôÄ ÎÖ∏Ïù¥Ï¶àÎ•º ÎçîÌï¥ ÎÖ∏Ïù¥Ï¶àÍ∞Ä ÏÑûÏù∏ Ïù¥ÎØ∏ÏßÄÎ•º ÎßåÎì†Îã§.

    # ÌùëÎ∞± Î≥ÄÌôò
    gray = cv2.cvtColor(noisy_img, cv2.COLOR_BGR2GRAY)
    #ÎÖ∏Ïù¥Ï¶àÍ∞Ä Ï∂îÍ∞ÄÎêú Ïª¨Îü¨ Ïù¥ÎØ∏ÏßÄÎ•º ÌùëÎ∞± Ïù¥ÎØ∏ÏßÄÎ°ú Î≥ÄÌôò

    # Îã§ÏñëÌïú ÌïÑÌÑ∞ Ï†ÅÏö©
    blurred = cv2.GaussianBlur(gray, (15, 15), 0) #15x15 Ïª§ÎÑêÎ°ú Í∞ÄÏö∞ÏãúÏïà Î∏îÎü¨(ÌùêÎ¶º) ÌïÑÌÑ∞Î•º Ï†ÅÏö©Ìï¥ ÎÖ∏Ïù¥Ï¶àÎ•º Ï§ÑÏûÖÎãàÎã§.
    edges = cv2.Canny(gray, 50, 150) #Ï∫êÎãà Ïó£ÏßÄ Í≤ÄÏ∂úÏùÑ ÏÇ¨Ïö©Ìï¥ Ïù¥ÎØ∏ÏßÄÏùò Ïú§Í≥ΩÏÑ†(Ïó£ÏßÄ)ÏùÑ Ï∂îÏ∂úÌï©ÎãàÎã§.

    # Í≤∞Í≥º Ï∂úÎ†•
    plt.figure(figsize=(15, 5)) #Ï†ÑÏ≤¥ Ï∂úÎ†• Í∑∏Î¶ºÏùò ÌÅ¨Í∏∞ ÏÑ§Ï†ï

    #ÎÖ∏Ïù¥Ï¶àÍ∞Ä Ï∂îÍ∞ÄÎêú ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÎ•º BGR‚ÜíRGBÎ°ú Î≥ÄÌôòÌï¥ Ï∂úÎ†•
    plt.subplot(1, 4, 1)
    plt.imshow(cv2.cvtColor(noisy_img, cv2.COLOR_BGR2RGB))
    plt.title('Original (with noise)')
    plt.axis('off')

    #ÌùëÎ∞± Ïù¥ÎØ∏ÏßÄÎ•º Ï∂úÎ†•
    plt.subplot(1, 4, 2)
    plt.imshow(gray, cmap='gray')
    plt.title('Grayscale')
    plt.axis('off')

    #Í∞ÄÏö∞ÏãúÏïà Î∏îÎü¨(ÌùêÎ¶º) Ï≤òÎ¶¨Îêú Ïù¥ÎØ∏ÏßÄÎ•º Ï∂úÎ†•
    plt.subplot(1, 4, 3)
    plt.imshow(blurred, cmap='gray')
    plt.title('Gaussian Blur')
    plt.axis('off')

    #Ïó£ÏßÄ(Ïú§Í≥ΩÏÑ†) Í≤ÄÏ∂ú Ïù¥ÎØ∏ÏßÄÎ•º Ï∂úÎ†•
    plt.subplot(1, 4, 4)
    plt.imshow(edges, cmap='gray')
    plt.title('Edge Detection')
    plt.axis('off')

    #ÏÑúÎ∏åÌîåÎ°Ø Í∞ÑÍ≤©ÏùÑ ÏûêÎèô Ï°∞Ï†ïÌïòÍ≥†, Ïù¥ÎØ∏ÏßÄÎ•º ÌôîÎ©¥Ïóê ÌëúÏãú
    plt.tight_layout()
    plt.show()

# Ìï®Ïàò Ïã§Ìñâ ÏΩîÎìú Ï∂îÍ∞Ä
if __name__ == "__main__":
    test4_image_filtering()
```

```python
# =============================================================================
# üìù ÌÖåÏä§Ìä∏ 5: Ïù¥ÎØ∏ÏßÄ ÏóÖÎ°úÎìú Î∞è Ï≤òÎ¶¨ (ÏÑ†ÌÉùÏÇ¨Ìï≠)
# =============================================================================

def test5_upload_and_process():  #Ìï®Ïàò Ï†ïÏùò
    """Ïù¥ÎØ∏ÏßÄ ÏóÖÎ°úÎìúÌïòÍ≥† Ï≤òÎ¶¨ÌïòÍ∏∞ (ÏÑ†ÌÉùÏÇ¨Ìï≠)"""
    print("\nüìÅ ÌÖåÏä§Ìä∏ 5: Ïù¥ÎØ∏ÏßÄ ÏóÖÎ°úÎìú Î∞è Ï≤òÎ¶¨")
    print("Ïù¥ÎØ∏ÏßÄ ÌååÏùºÏùÑ ÏóÖÎ°úÎìúÌïòÏÑ∏Ïöî...")

    #ÏòàÏô∏ Î∞úÏÉùÏùÑ ÎåÄÎπÑÌï¥ tryÎ∏îÎ°ù ÏãúÏûë
    try:
        # files.upload()Î•º ÌÜµÌï¥ ÏÇ¨Ïö©ÏûêÏóêÍ≤å ÌååÏùº ÏóÖÎ°úÎìú Ï∞ΩÏùÑ ÎùÑÏõÅÎãàÎã§
        uploaded = files.upload()

        if uploaded:
            filename = list(uploaded.keys())[0] #ÌååÏùºÏù¥ Ï†ïÏÉÅÏ†ÅÏúºÎ°ú ÏóÖÎ°úÎìú ÎêúÏßÄ ÌôïÏù∏ ÌõÑ ÌååÏùºÏùò Ïù¥Î¶ÑÏùÑ Í∞ÄÏ†∏Ïò¥.(Ïó¨Îü¨ ÌååÏùºÏù¥Î©¥ Ï≤´Î≤àÏû¨ ÌååÏùºÏùÑ Ïù¥Ïö©)

            # Ïù¥ÎØ∏ÏßÄ ÏùΩÍ∏∞
            image_data = uploaded[filename]
            img = Image.open(io.BytesIO(image_data))
            img_array = np.array(img)
            #ÏóÖÎ°úÎìúÎêú ÌååÏùºÏùò Ïù¥ÏßÑ Îç∞Ïù¥ÌÑ∞Î•º Î≥ÄÏàòÏóê Ï†ÄÏû•
            #Image.open(io.BytesIO(image_data))Î°ú Ïù¥ÎØ∏ÏßÄÎ•º Î©îÎ™®Î¶¨ÏóêÏÑú Î∞îÎ°ú ÏóΩÎãàÎã§.
            #Ïù¥ÎØ∏ÏßÄÎ•º NumPy Î∞∞Ïó¥Î°ú Î≥ÄÌôòÌï©ÎãàÎã§


            # RGB ‚Üí BGR Î≥ÄÌôò (OpenCVÏö©)
            if len(img_array.shape) == 3:
                img_bgr = cv2.cvtColor(img_array, cv2.COLOR_RGB2BGR)
                #Ïª¨Îü¨Ïù¥ÎØ∏ÏßÄ(3Ï±ÑÎÑê)Ïùò Í≤ΩÏö∞ OpenCVÏóêÏÑú ÏÇ¨Ïö©ÌïòÎäî BGRÏàúÏÑúÎ°ú Î≥ÄÌôò
            else:
                img_bgr = img_array
                #ÌùëÎ∞±Ïù¥ÎØ∏ÏßÄ(2Ï±ÑÎÑê)Ïù∏ Í≤ΩÏö∞Îäî Î≥ÄÌôò ÏóÜÏù¥ Í∑∏ÎåÄÎ°ú ÏÇ¨Ïö©

            # ÌÅ¨Í∏∞ Ï°∞Ï†ï (ÎÑàÎ¨¥ ÌÅ∞ Ïù¥ÎØ∏ÏßÄ Î∞©ÏßÄ)
            height, width = img_bgr.shape[:2]
            if width > 800:  #ÎÑàÎπÑÍ∞Ä 800ÌîΩÏÖÄÏùÑ ÎÑòÏúºÎ©¥ ÎπÑÏú®ÏùÄ Ïú†ÏßÄÌïú Ï±Ñ ÎÑàÎπÑÎ•º 800ÏúºÎ°ú Ï§ÑÏù∏Îã§
                scale = 800 / width
                new_width = 800
                new_height = int(height * scale)
                img_bgr = cv2.resize(img_bgr, (new_width, new_height))

            # Îã§ÏñëÌïú Ï≤òÎ¶¨ Ï†ÅÏö©
            gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY) #ÌùëÎ∞± Î≥ÄÌôò
            blurred = cv2.GaussianBlur(img_bgr, (21, 21), 0) #Î∏îÎü¨ Ï≤òÎ¶¨
            edges = cv2.Canny(gray, 100, 200) #Ïó£ÏßÄ Í≤ÄÏ∂ú

            # Í≤∞Í≥º Ï∂úÎ†•
            plt.figure(figsize=(20, 5)) #Ï†ÑÏ≤¥ Ï∂úÎ†• Í∑∏Î¶ºÏùò ÌÅ¨Í∏∞ ÏÑ§Ï†ï

            #ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÎ•º BGR->RGBÎ°ú Î≥ÄÌôò ÌõÑ Ï∂úÎ†•
            plt.subplot(1, 4, 1)
            plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
            plt.title('Original')
            plt.axis('off')

            #ÌùëÎ∞± Ïù¥ÎØ∏ÏßÄ Ï∂úÎ†•
            plt.subplot(1, 4, 2)
            plt.imshow(gray, cmap='gray')
            plt.title('Grayscale')
            plt.axis('off')

            #Î∏îÎü¨Îêú Ïù¥ÎØ∏ÏßÄÎ•º BGR‚ÜíRGBÎ°ú Î≥ÄÌôòÌï¥ Ï∂úÎ†•
            plt.subplot(1, 4, 3)
            plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
            plt.title('Blurred')
            plt.axis('off')

            #Ïó£ÏßÄ(Ïú§Í≥ΩÏÑ†) Í≤ÄÏ∂ú Ïù¥ÎØ∏ÏßÄÎ•º Ï∂úÎ†•
            plt.subplot(1, 4, 4)
            plt.imshow(edges, cmap='gray')
            plt.title('Edge Detection')
            plt.axis('off')


            #ÏÑúÎ∏åÌîåÎ°Ø Í∞ÑÍ≤©ÏùÑ ÏûêÎèô Ï°∞Ï†ïÌïòÍ≥†, Ïù¥ÎØ∏ÏßÄÎ•º ÌôîÎ©¥Ïóê ÌëúÏãú
            plt.tight_layout()
            plt.show()

            print(f"‚úÖ Ïù¥ÎØ∏ÏßÄ Ï≤òÎ¶¨ ÏôÑÎ£å! ÌÅ¨Í∏∞: {img_bgr.shape}")

    except Exception as e:
        print(f"‚ùå ÏóÖÎ°úÎìú Ï§ë Ïò§Î•ò: {e}")
        print("Ïù¥ ÌÖåÏä§Ìä∏Îäî Í±¥ÎÑàÎõ∞Í≤†ÏäµÎãàÎã§.")

# Ìï®Ïàò Ïã§Ìñâ ÏΩîÎìú Ï∂îÍ∞Ä
if __name__ == "__main__":
   test5_upload_and_process()
```
